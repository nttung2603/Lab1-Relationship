# Tiền xử lý

```python
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx', sheet_name='country')
```

```python
df.head()
```

```python
df.drop('#', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
num_cols=['Total Cases', 'New Cases', 'Total Deaths', 
          'New Death', 'Total Recovered', 'Active Case', 
          'Serious Critical', 'Tot Cases/1M pop', 'Death/1M pop', 
          'Total test', 'Test/1M pop', 'Population']

for col in num_cols:
    df[col]=df[col].str.replace(',', '')
    df[col]=pd.to_numeric(df[col])
```

```python
df.dtypes
```
